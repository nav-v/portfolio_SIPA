## Topic 7: Merging DataFrames

### Concept
Combine two dataframes based on common columns.

**General Pattern:**
```python
pd.merge(df1, df2, on='CommonColumn', how='inner')
pd.merge(df1, df2, left_on='Col1', right_on='Col2', how='left')
```

**Types of merges:**
- `inner`: Only matching rows from both
- `outer`: All rows from both (fills NaN where no match)
- `left`: All from left, matching from right
- `right`: All from right, matching from left
