### ðŸ“‹ Guidelines (Cheat Sheet)

## Topic 2: np.where (Nested Logic)

### Concept
You can put an `np.where` **inside** another `np.where` to handle 3+ categories (High/Med/Low).

**Syntax:**
```python
np.where(condition1, 'Value1',
    np.where(condition2, 'Value2', 'Value3')
)
```

## Topic 5: Pivot Tables (Drill)

### Concept
Turning "Long" data (many rows) into "Wide" data (many columns).

**Syntax:**
```python
df.pivot(
    index='WhatStaysAsRows',
    columns='WhatBecomesColumns',
    values='WhatFillsCells'
)
```
