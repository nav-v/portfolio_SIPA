**Q7:** `products[products['ProdName'].str.contains('Pro')]`

**Q8:** `customers['AgeMonths'] = customers['Age'] * 12`

**Q9:**
```python
products['Price'] = products['Price'].str.replace('$', '').str.replace(',', '')
products['Price'] = pd.to_numeric(products['Price'], errors='coerce').fillna(0)
```

**Q10:**
```python
products['PriceCategory'] = np.where(products['Price'] > 1000, 'High',
                            np.where(products['Price'] > 100, 'Medium', 'Low'))
```

**Q11:** `merged_df = pd.merge(orders, customers, on='CustID')`

**Q12:** `pd.merge(customers, orders, on='CustID', how='left')`

**Q13:** `full_data = pd.merge(merged_df, products, on='ProdID')`

**Q14:** `monthly_stats.melt(id_vars='Region', var_name='Month', value_name='Sales')`

**Q15:** `melted_df.pivot(index='Region', columns='Month', values='Sales')`

**Q16:** `full_data.groupby('City')['Qty'].sum()`

**Q17:** `full_data.groupby('Category').agg({'Qty': 'sum', 'Price': 'mean'})`

**Q18:** `px.bar(full_data.groupby('Category')['Qty'].sum().reset_index(), x='Category', y='Qty')`

**Q19:** `px.scatter_geo(locations, lat='Lat', lon='Lon')`

**Q20:**
```python
orders.merge(products, on='ProdID') \
      .loc[lambda x: x['Category'] == 'Electronics'] \
      .groupby('ProdName')['Qty'].sum()
```
